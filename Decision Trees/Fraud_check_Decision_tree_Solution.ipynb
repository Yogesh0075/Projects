<b>Use decision trees to prepare a model on fraud data 
treating those who have taxable_income <= 30000 as "Risky" and others are "Good"

<b>Data Description :

<b>Undergrad : person is under graduated or not
Marital.Status : marital status of a person
Taxable.Income : Taxable income is the amount of how much tax an individual owes to the government 
Work Experience : Work experience of an individual person
Urban : Whether that person belongs to urban area or not

```python
#Converting categorical variables into numerical variables
cat_var = pd.get_dummies(data_2[['Undergrad','Urban']])
cat_var
```

```python
from sklearn.preprocessing import OneHotEncoder

enc = OneHotEncoder()
enc_data = enc.fit_transform(data_2['Marital.Status'].values.reshape(-1,1)).toarray()
enc_df = pd.DataFrame(enc_data,columns=enc.get_feature_names_out(['Marital.Status']))
data_3 = pd.concat([data_2,enc_df],axis=1)
data_3
```

```python
data_4 = data_3.drop(['Marital.Status'],axis=1)
data_4
```

```python
data_5 = pd.concat([data_4,cat_var],axis=1)
data_5
```

```python
data_6 = data_5.drop(['Undergrad','Urban'],axis=1)
data_6
```

```python
x = data_6.drop(['Taxable.Income'],axis=1)
y = data_6['Taxable.Income']
```

<b>Model Building with Entropy criterion

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=101)
model_1 = DecisionTreeClassifier(criterion='entropy',max_depth=7)
```

```python
model_1.fit(x_train,y_train)
```

```python
fn = ['City.Population','Work.Experience','Marital.Status_Divorced','Marital.Status_Married','Marital.Status_Single','Undergrad_NO','Undergrad_YES','Urban_NO','Urban_YES']
cn = ['Risky','Good']
plot_tree(model_1,feature_names=fn,class_names=cn,filled=True);
```

```python
y_pred = model_1.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix,classification_report,roc_curve,roc_auc_score
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

```python
model_1.score(x_train,y_train)
```

```python
model_1.score(x_test,y_test)
```

```python
from sklearn.preprocessing import LabelEncoder
yt = LabelEncoder().fit_transform(y_test)
yp = LabelEncoder().fit_transform(y_pred)
```

```python
fpr , tpr , threshold = roc_curve(yt,model_1.predict_proba(x_test)[:,1])
auc = roc_auc_score(yt,yp)
plt.figure(figsize=(5,5))
plt.plot(fpr,tpr,color='b',linestyle='--')
plt.plot([0,1],[0,1],'k--')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.show()
```

```python
auc
```

<b>Lets us Introduce SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

x_smote , y_smote = smote.fit_resample(x , y)
```

```python
x_smote
```

```python
y_smote.value_counts()
```

```python
y.value_counts()
```

<b>Let us also standardize the data for better resullts

```python
#Let us also standardize the data for better results
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_scaled = scaler.fit_transform(x_smote)
x_scaled
```

```python
x1_train,x1_test,y1_train,y1_test = train_test_split(x_scaled,y_smote,test_size=0.2,random_state=102)
model_3 = DecisionTreeClassifier(criterion='entropy',max_depth=7)
```

```python
model_3.fit(x1_train,y1_train)
```

```python
plt.figure(dpi=500)
plot_tree(model_3,feature_names=fn,class_names=cn,filled=True);
plt.show()
```

```python
y_pred_3 = model_3.predict(x1_test)
y_pred_3
```

```python
print(confusion_matrix(y1_test,y_pred_3))
```

```python
print(classification_report(y1_test,y_pred_3))
```

```python
y1t = LabelEncoder().fit_transform(y1_test)
yp3 = LabelEncoder().fit_transform(y_pred_3)
```

```python
fpr , tpr , threshold = roc_curve(y1t,model_3.predict_proba(x1_test)[:,1])
auc = roc_auc_score(y1t,yp3)
plt.figure(figsize=(5,5))
plt.plot(fpr,tpr,color='b',linestyle='--')
plt.plot([0,1],[0,1],'k--')
plt.xlabel("False Positive rate")
plt.ylabel("Trur Positive rate")
plt.show()
```

```python
auc
```
